```python
import networkx as nx 
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

# SBM Network

```python
model_type='SBM'
sizes = [400, 400, 1200]
probs = [[0.4, 0.01, 0.01],
         [0.01, 0.6, 0.01],
         [0.01, 0.01, 0.9]]
SBM = nx.stochastic_block_model(sizes, probs, seed=0)
A=nx.to_numpy_matrix(SBM)
```

```python
plt.rcParams["figure.figsize"] = (7,7)
plt.matshow(A)
plt.colorbar()
plt.show()
```

# Mixed Membership Model

```python
import graspologic

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from graspologic.simulations import mmsbm

rng = np.random.default_rng(123)

#number of nodes
n = 1000

#Community Probability Matrix
#C = [[0.7, 0.2,0.1],
     #[0.2, 0.05,0.3],
    #[0.1,0.3,0.4]]
    
C = [[1.0, 0.2,0.4],
     [0.2, 1,0.1],
    [0.4,0.1,1]]

#Parameter for Dir distribution
alpha = [100,100,1000]
#alpha = [0.1]*4

G = mmsbm(n, C, alpha, rng = rng)
```

```python
from graspologic.plot import heatmap

heatmap(G, cbar= False, title ='MMSBM Simulation');
```
